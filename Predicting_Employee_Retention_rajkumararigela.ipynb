7.2.3 Fit logistic regression model <font color="red">[3 Marks]</font>

```python
# Fit a logistic regression model on X_train after adding a constant and output the summary
# Select only the columns selected by RFE, drop Attrition column from features
x_train_selected = x_train[col].drop('Attrition', axis=1)

# Convert boolean and integer columns to float
x_train_selected = x_train_selected.astype(float)

# Add a constant to the training set
x_train_constant = sm.add_constant(x_train_selected)

# Ensure the DataFrame with constant is of float type
x_train_constant = x_train_constant.astype(float)

# Fit a logistic regression model on X_train after adding a constant and output the summary
logm2=sm.GLM(y_train,x_train_constant,family=sm.families.Binomial())
res=logm2.fit()
res.summary()
```

**Model Interpretation**

The output summary table  will provide the features used for building model along with coefficient of each of the feature and their p-value. The p-value in a logistic regression model is used to assess the statistical significance of each coefficient. Lesser the p-value, more significant the feature is in the model.

A positive coefficient will indicate that an increase in the value of feature would increase the odds of the event occurring. On the other hand, a negative coefficient means the opposite, i.e,  an increase in the value of feature would decrease the odds of the event occurring.



7.2.4 Evaluate VIF of features <font color="red">[3 Marks]</font>

```python
# Import 'variance_inflation_factor'
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Make a VIF DataFrame for all the variables present
vif = pd.DataFrame()
vif['Features'] = x_train_selected.columns
vif['VIF'] = [variance_inflation_factor(x_train_selected.values, i) for i in range(x_train_selected.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# Define ROC function
def plot_roc_curve(atual, predicted):
    fpr, tpr, thresholds = metrics.roc_curve(atual, predicted,drop_intermediate= False)
    accuracy_score=metrics.roc_auc_score(atual, predicted)
    plt.plot(fpr, tpr, color='orange', label='ROC curve(area =%0.2f)'% accuracy_score)
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
```
